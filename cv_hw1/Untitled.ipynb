```python
#x must be a numpy array
def psi(x,offset,sigma):
    c2 = sum(np.exp(2*np.pi*1j*x/4)*np.exp((-x**2)/2)) / sum(np.exp((-x**2)/2))
    func = (1/np.sqrt(sigma)) * (np.exp((2*np.pi*1j)*(x-offset)/(4*sigma)) - c2) * (np.exp((-(x-offset)**2)/(2*sigma**2)))
    func_star = np.conj(func)
    c1 = np.sqrt(1./sum(func*func_star))
    func = c1*func
    return func
```

```python
x_range = np.arange(-10,10,.1)
#x_range = np.arange(0,img[0].size,1)
plt.plot(x_range,psi(x_range,0,1))
plt.show()
```
